```python
import random

def jogar_adivinhacao():
    """
    FunÃ§Ã£o principal do Jogo de AdivinhaÃ§Ã£o com tema Halloween.
    A bruxa guarda um nÃºmero secreto em seu caldeirÃ£o.
    """

    # 1. ConfiguraÃ§Ã£o do Jogo
    print("ğŸƒ Bem-vindo Ã  Cabana da Bruxa! ğŸ§™â€â™€ï¸")
    print("Para sair das garras da Bruxa, descubra o 'NÃºmero Secreto' em seu caldeirÃ£o.")
    print("VocÃª tem poucas chances para descobrir, ou *UM FEITIÃ‡O* cairÃ¡ sobre vocÃª!\n")

    # Gera um nÃºmero secreto aleatÃ³rio entre 1 e 50
    # (A Componente 'random' serÃ¡ usada para geraÃ§Ã£o de nÃºmeros pseudo-aleatÃ³rios.)
    numero_secreto = random.randint(1, 50)

    # Define o nÃºmero de tentativas
    tentativas_maximas = 5
    # Controle: VariÃ¡vel de controle que usarÃ¡ o 'break' para sair do loop.
    acertou = False
    tentativas_restantes = tentativas_maximas

    # 2. InÃ­cio do Jogo
    # O conceito: 'while' loop para controlar o fluxo do jogo enquanto o jogador tiver chances e nÃ£o acertar.
    while tentativas_restantes > 0 and not acertou:
        print(f"\nâœ¨ Tentativas restantes: {tentativas_restantes}")

        # Leitura da entrada do usuÃ¡rio
        try:
            chute = int(input("ğŸ’€ Qual o seu palpite entre 1 e 50? "))
        except ValueError:
            print("âš ï¸ Entrada invÃ¡lida! Digite apenas nÃºmeros inteiros.\n")
            continue  # Volta para o inÃ­cio do loop sem decrementar tentativas

        # 3. ValidaÃ§Ã£o da Entrada
        if chute < 1 or chute > 50:
            print("ğŸš« O palpite deve estar entre 1 e 50. Preste atenÃ§Ã£o!")
            continue

        # 4. VerificaÃ§Ã£o do Palpite (Estruturas Condicionais)
        tentativas_restantes -= 1

        if chute == numero_secreto:
            print(f"ğŸ‰ ParabÃ©ns! VocÃª quebrou o feitiÃ§o! O nÃºmero era {numero_secreto}!")
            acertou = True
        elif chute > numero_secreto:
            print("â˜ ï¸ Muito alto! O caldeirÃ£o borbulha dizendo que o nÃºmero Ã© *MENOR*...\n")
        else:
            print("ğŸ’€ Muito baixo! A fumaÃ§a do caldeirÃ£o indica que o nÃºmero Ã© *MAIOR*...\n")

    # 5. Fim do Jogo
    if not acertou:
        print("ğŸ‘» Suas chances acabaram! O feitiÃ§o estÃ¡ completo!")
        print(f"O NÃºmero Secreto da Bruxa era {numero_secreto}.")

if __name__ == "__main__":
    jogar_adivinhacao()
```

```python
import random
import os  # MÃ³dulo para limpar a tela do console

def exibir_forca(erros):
    """Desenha a forca de acordo com o nÃºmero de erros."""
    # A forca Ã© um exemplo de como o algoritmo pode representar estados.
    stages = [
        """
        ------
        |    |
        |
        |
        |
        |
        =========
        """,
        # 1 erro
        """
        ------
        |    |
        |    O
        |
        |
        |
        =========
        """,
        # 2 erros
        """
        ------
        |    |
        |    O
        |    |
        |
        |
        =========
        """,
        # 3 erros
        """
        ------
        |    |
        |    O
        |   /|
        |
        |
        =========
        """,
        # 4 erros
        """
        ------
        |    |
        |    O
        |   /|\\
        |
        |
        =========
        """,
        # 5 erros (CORPO COMPLETO)
        """
        ------
        |    |
        |    O
        |   /|\\
        |   / \\
        |
        =========
        """
    ]
    print(stages[erros])

def jogar_forca():
    """FunÃ§Ã£o principal do Jogo da Forca com tema Halloween."""
    print("ğŸ•¸ï¸ğŸƒ Bem-vindo ao Jogo da Forca de Halloween! ğŸ§™â€â™€ï¸\n")
    print("Tente adivinhar a palavra MISTERIOSA (relacionada ao tema Halloween)")
    print("Mas cuidado... a cada erro a bruxa se aproxima atÃ© que o *FEITIÃ‡O* se complete!\n")

    # 1. ConfiguraÃ§Ã£o (Lista de Palavras e Sorteio)
    palavras_tema = ["ABÃ“BORA", "BRUXA", "MAGO", "VASSOURA", "FANTASMA", "ASSOMBRAÃ‡ÃƒO",
                     "CAVEIRA", "FEITIÃ‡O", "ESQUELETO", "CALDEIRÃƒO", "MURCHO", "CONJURO"]
    palavra_secreta = random.choice(palavras_tema).upper()

    # 2. VariÃ¡veis de controle
    letras_descobertas = ["_" for letra in palavra_secreta]
    letras_erradas = []
    erros = 0
    max_erros = 5

    # 3. Loop Principal
    while erros < max_erros and "_" in letras_descobertas:
        os.system("cls" if os.name == "nt" else "clear")
        exibir_forca(erros)
        print("\nPalavra: ", " ".join(letras_descobertas))
        print(f"ğŸ©¸ Letras erradas: {', '.join(letras_erradas)}")
        print(f"Tentativas restantes: {max_erros - erros}\n")

        # Entrada do jogador
        chute = input("ğŸ”® Escolha uma letra: ").upper()

        if len(chute) != 1 or not chute.isalpha():
            print("âš ï¸ Digite apenas uma letra vÃ¡lida!\n")
            continue

        # 4. VerificaÃ§Ã£o da letra
        if chute in letras_descobertas or chute in letras_erradas:
            print("âš ï¸ VocÃª jÃ¡ tentou essa letra! Tente outra.\n")
            continue

        if chute in palavra_secreta:
            print("âœ¨ Boa! A letra faz parte da palavra.\n")
            for i, letra in enumerate(palavra_secreta):
                if letra == chute:
                    letras_descobertas[i] = chute
        else:
            print("ğŸ’€ A bruxa riu! Essa letra nÃ£o estÃ¡ na palavra...\n")
            letras_erradas.append(chute)
            erros += 1

    # 5. Encerramento do jogo
    os.system("cls" if os.name == "nt" else "clear")
    exibir_forca(erros)

    if "_" not in letras_descobertas:
        print(f"ğŸ‰ ParabÃ©ns! VocÃª quebrou o feitiÃ§o e descobriu a palavra: {palavra_secreta}!")
    else:
        print("ğŸ‘» O feitiÃ§o foi lanÃ§ado! A bruxa venceu desta vez...")
        print(f"A palavra secreta era: {palavra_secreta}")

if __name__ == "__main__":
    jogar_forca()
```

```python
import random
import time
import os

def limpar_tela():
    os.system('cls' if os.name == 'nt' else 'clear')

def introducao():
    limpar_tela()
    print("ğŸšï¸ Bem-vindo Ã  MansÃ£o Assombrada! ğŸƒ")
    time.sleep(1)
    print("VocÃª Ã© um caÃ§ador de monstros e precisa capturar o ser terrÃ­vel que se esconde aqui...")
    time.sleep(2)
    print("Mas cuidado... hÃ¡ armadilhas, sustos e criaturas te observando nas sombras. ğŸ‘€\n")
    time.sleep(2)

def criar_mapa():
    """Cria uma grade 3x3 da mansÃ£o."""
    return [["ğŸ•¯ï¸" for _ in range(3)] for _ in range(3)]

def exibir_mapa(mapa):
    for linha in mapa:
        print(" ".join(linha))
    print()

def jogar_caca_ao_monstro():
    introducao()

    mapa = criar_mapa()
    monstro_x = random.randint(0, 2)
    monstro_y = random.randint(0, 2)
    jogador_x, jogador_y = 1, 1  # comeÃ§a no centro
    tentativas = 5

    print("ğŸ—ï¸ Dica: O monstro estÃ¡ escondido em uma das 9 salas da mansÃ£o!")
    print("Use W (cima), S (baixo), A (esquerda), D (direita) para se mover.\n")

    while tentativas > 0:
        print(f"ğŸ’€ Tentativas restantes: {tentativas}\n")
        exibir_mapa(mapa)

        movimento = input("ğŸ‘‰ Seu movimento (W/A/S/D): ").strip().upper()
        if movimento not in ["W", "A", "S", "D"]:
            print("âš ï¸ Movimento invÃ¡lido! Tente novamente.\n")
            continue

        if movimento == "W" and jogador_x > 0:
            jogador_x -= 1
        elif movimento == "S" and jogador_x < 2:
            jogador_x += 1
        elif movimento == "A" and jogador_y > 0:
            jogador_y -= 1
        elif movimento == "D" and jogador_y < 2:
            jogador_y += 1
        else:
            print("ğŸšª VocÃª bateu numa parede da mansÃ£o! Tente outra direÃ§Ã£o.\n")
            continue

        limpar_tela()

        print("ğŸ‘£ VocÃª caminha... o chÃ£o range sob seus pÃ©s...")
        time.sleep(1)

        if jogador_x == monstro_x and jogador_y == monstro_y:
            print("ğŸ˜± O monstro aparece! VocÃª o capturou com sua cruz sagrada! âœï¸")
            print("ğŸ‰ ParabÃ©ns! A aldeia estÃ¡ a salvo novamente!")
            break
        else:
            distancia = abs(monstro_x - jogador_x) + abs(monstro_y - jogador_y)
            if distancia == 1:
                print("ğŸ’¨ VocÃª sente um cheiro terrÃ­vel... o monstro estÃ¡ *perto!* ğŸ‘€\n")
            elif distancia == 2:
                print("ğŸ‘‚ VocÃª ouve um sussurro distante... algo se move nas sombras...\n")
            else:
                print("ğŸŒ«ï¸ SilÃªncio total... talvez o monstro esteja longe.\n")

        tentativas -= 1

    if tentativas == 0:
        print("ğŸ’€ O monstro te encontrou primeiro! A mansÃ£o ecoa com seus gritos... ğŸ©¸")
        print("ğŸ•¯ï¸ Fim de jogo.")

    print("\nDeseja jogar novamente? (S/N)")
    if input("> ").strip().upper() == "S":
        jogar_caca_ao_monstro()
    else:
        print("\nğŸƒ A mansÃ£o se apaga... e vocÃª desaparece na noite...")

if __name__ == "__main__":
    jogar_caca_ao_monstro()
```
